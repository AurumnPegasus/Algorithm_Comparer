```python
import matplotlib.pyplot as plt 
```

```python
def readFiles(filename):
    f = open(filename, 'r+')
    lines = f.readlines()
    time = []
    for line in lines:
        time.append(float(line.strip()))
    return time
```

```python
def plotGraphs(x, y, title):
    plt.plot(x, y)
    plt.xlabel('size of the array') 
    plt.ylabel('time taken') 
    plt.title(title) 
    plt.show() 
    return
```

```python
timeA = readFiles('A.txt')
timeB = readFiles('B.txt')
N = 5
x = []
for i in range(1, 6):
    x.append(i)
plotGraphs(x, timeA, "Merge Sort")
plotGraphs(x, timeB, "3 way Merge Sort")
```
