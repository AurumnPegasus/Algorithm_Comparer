```python
import matplotlib.pyplot as plt 
```

```python
def readFiles(filename):
    f = open(filename, 'r+')
    lines = f.readlines()
    time = []
    for line in lines:
        time.append(float(line.strip()))
    return time
```

```python
def plotGraphs(x, y1, y2, title):
    plt.plot(x, y1)
    plt.plot(x, y2)
    plt.xlabel('size of the array') 
    plt.ylabel('time taken') 
    plt.title(title)
    plt.legend(["Recursive Merge Sort", "Recursive 3-Way Merge sort"])
    plt.show() 
    return
```

```python
timeA = readFiles('A.txt')
timeB = readFiles('B.txt')
N = 5
x = []
for i in range(1, 6):
    x.append(i)
plotGraphs(x, timeA, timeB, "Merge Sort")
```
